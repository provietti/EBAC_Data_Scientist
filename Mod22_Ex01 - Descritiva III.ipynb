#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
#importando as bilbiotecas necessárias
import pandas as pd
import plotly.express as px
import yfinance as yf
from pandas_datareader import data as web
```

```python
%%time
#baixando os dados de três ativos do Yahoo Finance para um dataframe, a partir de 01/01/2020. 
ativos = ['ELET3.SA', 'AMER3.SA', 'ABEV3.SA']
dt_ini = '2020-01-01'
yf.pdr_override()
df = web.get_data_yahoo(ativos, start=dt_ini)
df.head()
```

```python
#montando dataframe para a criação do gráfico, considerando os valores ajustados no fechamento('Adj Close')
fechamento_vertical = df['Adj Close'].stack().reset_index()
fechamento_vertical.columns = ['data', 'ativo', 'valor']
#criando gráfico do plotly
fig = px.line(fechamento_vertical, x="data", y="valor", color='ativo', template='plotly_white')
fig.show()
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Montando DataFrame com a amplitude de variação do valor do ativo dentro do dia
df_amplitude = pd.DataFrame()
df_amplitude['ABEV3.SA'] = df[('High','ABEV3.SA')] -  df[('Low','ABEV3.SA')]
df_amplitude['AMER3.SA'] = df[('High','AMER3.SA')] -  df[('Low','AMER3.SA')]
df_amplitude['ELET3.SA'] = df[('High','ELET3.SA')] -  df[('Low','ELET3.SA')]
df_amplitude = df_amplitude.stack().reset_index()
df_amplitude.columns = ['Data','Ativo','Amplitude']
# Montando o gráfico  das amplitudes com o plotly
fig = px.line(df_amplitude, x="Data", y="Amplitude", color='Ativo', template='plotly_white')
fig.show()
```

Até o início de 2022, entre os ativos analisados, o AMER3.SA possuia a maior volatilidade. Atualmente o ativo com maior volatilidade é a ELET3.SA
